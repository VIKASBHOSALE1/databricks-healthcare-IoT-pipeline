```python
%sql
-- Write a query to get avg_heart_rate by shift and day
-- from patient_vitals_hourly
-- order by day ascending
-- This will be plotted as a multi-line chart in the dashboard
select date, shift, avg(avg_heart_rate) as avg_heart_rate from cat_healthcare_iot_patients.gold.gold_shift_summary group by date, shift order by shift,date asc
```

```python
%sql
-- Write a query to count alerts grouped by date
-- from anomaly_alerts
-- order by count descending
--Bar chart
select date, count(*) as `Alerts` from cat_healthcare_iot_patients.gold.gold_anomaly_alerts group by date order by `Alerts` desc
```

```python
%sql
-- Write a query to show the last 20 critical alerts
-- from anomaly_alerts
-- where alert_severity = 'CRITICAL'
-- order by alert_time descending
-- select: patient_id, ward_id, alert_time, heart_rate, spo2, temperature, alert_category
--Table
select Patient_ID, date, Heart_Rate_bpm, Temperature_C, alert_severity, is_anomaly from cat_healthcare_iot_patients.gold.gold_anomaly_alerts where alert_severity = 'CRITICAL' order by Silver_Processed_At desc limit 20
```
